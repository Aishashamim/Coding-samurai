```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data Loading
file_path = 'your_dataset.csv'
data = pd.read_csv(r"C:\Users\a\Downloads\calendar.csv")

# Step 2: Data Exploration
print("First few rows of the dataset:")
print(data.head())

# overview of the dataset
print("\nDataset info:")
print(data.info())

# Summary statistics
print("\nSummary statistics:")
print(data.describe())

# Step 3: Data Cleaning
data.dropna(inplace=True) 

# Step 4: Basic Statistics
# Average price of listings
average_price = data['price'].mean()
print("\nAverage price of listings:", average_price)

# Distribution of property types
plt.figure(figsize=(10, 6))
property_type_counts = data['property_type'].value_counts()
property_type_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Property Types')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Distribution of neighborhoods
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='neighborhood', order=data['neighborhood'].value_counts().index, palette='viridis')
plt.title('Distribution of Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Step 5: Visualization
# Example: Price distribut
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=30, kde=True, color='purple')
plt.title('Price Distribution of Listings')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
